```python
df_penguin = pd.read_csv('penguins_simple.csv')
df_penguin.dropna(inplace=True)
df_bill_length = df_penguin['bill_length_mm']

# print min bill length
print("Chiu di m nh nht:", df_bill_length.min())
# print max bill length
print("Chiu di m ln nht:", df_bill_length.max())
# print mean bill length
print("Chiu di m trung bnh:", df_bill_length.mean())
# print median bill length
print("Chiu di m trung v:", df_bill_length.median())

plt.figure(figsize=(10, 6))
sns.boxplot(y='bill_length_mm', data=df_penguin)
plt.show()
```

```python
df_sub = df_penguin[['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g']]
print(df_sub.head())
```

```python
corr_matrix = df_sub.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', square=True)
plt.show()
```

```python
ngay = pd.date_range(start='2025-06-11', periods=30, freq='D')

nhiet_do = np.random.uniform(25, 39, size=30).astype(int)
do_am = np.random.uniform(60, 90, size=30).astype(int)
luong_mua = np.random.uniform(0, 10, size=30).astype(int)

plt.figure(figsize=(12, 6))

plt.plot(ngay, nhiet_do, marker='o', color='red')
plt.plot(ngay, do_am, marker='s', color='blue')
plt.plot(ngay, luong_mua, marker='^', color='green')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.subplot(3, 1, 1)
plt.plot(ngay, nhiet_do, marker='o', color='red')
plt.title('Nhit  trung bnh hng ngy')

plt.subplot(3, 1, 2)
plt.plot(ngay, do_am, marker='s', color='blue')
plt.title(' m trung bnh hng ngy')

plt.subplot(3, 1, 3)
plt.plot(ngay, luong_mua, marker='^', color='green')
plt.title('Lng ma trung bnh hng ngy')

plt.show()
```

```python
df_weather = pd.DataFrame({
    'Date': ngay,
    'Temp': nhiet_do,
    'Humid': do_am,
    'Rain': luong_mua
})
corr_matrix = df_weather.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.show()
```
