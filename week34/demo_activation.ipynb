```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.layers import Dense, Activation, Dropout
from tensorflow.keras.models import Sequential
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
x = np.linspace(-10, 10, 100)
y_sigmoid = sigmoid(x)

plt.plot(x, y_sigmoid, label='Sigmoid', color='blue')
plt.title('Activation Function: Sigmoid')
plt.xlabel('Input')
plt.ylabel('Output')
plt.grid()
plt.legend()
plt.show()
```

```python
def tanh(x):
    return np.tanh(x)

y_tanh = tanh(x)
plt.plot(x, y_tanh, label='Tanh', color='orange')
plt.title('Activation Function: Tanh')
plt.xlabel('Input')
plt.ylabel('Output')
plt.grid()
plt.legend()
plt.show()
```

```python
def relu(x):
    return np.maximum(0, x)

plt.figure(figsize=(12, 6))
y_relu = relu(x)
plt.plot(x, y_relu, label='ReLU', color='green')
plt.title('Activation Function: ReLU')
plt.xlabel('Input')
plt.ylabel('Output')
plt.grid()
plt.legend()
plt.show()
```

```python
def softmax(x):
    e_x = np.exp(x - np.max(x))
    return e_x / e_x.sum(axis=0)
```

```python
x = [-2, -1, 0, 1, 2]
y = softmax(x)
print("Softmax Output:", y)
print(sum(y))  # Should be 1.0
```

```python
x1 = np.linspace(-10, 10, 100)
x2 = x1 + np.random.normal(0, 1, 100)
x3 = x1 + np.random.normal(0, 1, 100)
y = softmax(np.array([x1, x2, x3]))
print("Softmax Output (3 classes):\n", y[:, :3])  
print("Sum across classes (should be 1):", np.sum(y[:, :3], axis=0))  # Should be 1.0 for each sample
```
