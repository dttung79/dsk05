```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.layers import Dense, Activation
from tensorflow.keras.models import Sequential
```

```python
X = np.linspace(0, 10, 100)
Y = 2.5 * X + 1 + np.random.normal(0, 2, 100)
plt.figure(figsize=(8, 6))
plt.scatter(X, Y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter plot of X vs Y')
plt.grid()
plt.show()
```

```python
model = Sequential([
    Dense(1, input_shape=(1,), activation=None)  # No activation for linear regression
])

model.compile(optimizer='sgd', loss='mean_squared_error')

# train model
model.fit(X, Y, epochs=100, verbose=1)
```

```python
loss_values = model.history.history['loss']
plt.figure(figsize=(8, 6))
plt.plot(loss_values)
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Training Loss Over Epochs')
plt.grid()
plt.show()
```

```python
Y_pred = model.predict(X)
plt.figure(figsize=(8, 6))
plt.scatter(X, Y, label='Data', color='blue')
plt.scatter(X, Y_pred, label='Model Prediction', color='red')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter plot of X vs Y with Model Prediction')
plt.legend()
plt.grid()
plt.show()
```

```python
weights, biases = model.layers[0].get_weights()
print("Weights:\n", weights)
print("Biases:\n", biases)
```

```python
model = Sequential([
    Dense(30, input_shape=(1, ), activation='tanh'),
    Dense(30, activation='tanh'),
    Dense(1, activation=None)   # Output layer for regression
])
```

```python
model.compile(optimizer='sgd', loss='mean_squared_error')
model.fit(X, Y, epochs=100, verbose=0)

loss_values = model.history.history['loss']
plt.figure(figsize=(8, 6))
plt.plot(loss_values)
plt.xlabel('Epoch') 
plt.ylabel('Loss')
plt.title('Training Loss Over Epochs for Deep NN')
plt.grid()
plt.show()
```

```python
Y_pred = model.predict(X)
plt.figure(figsize=(8, 6))
plt.scatter(X, Y, label='Data', color='blue')
plt.scatter(X, Y_pred, label='Model Prediction', color='red')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter plot of X vs Y with Deep NN Model Prediction')
plt.legend()
plt.grid()
plt.show()
```

```python
X = np.linspace(-20, 20, 500)
Y = 2.0 * X ** 2 - 3.0 * X + 5.0 + np.random.normal(0, 20, 500)
plt.figure(figsize=(8, 6))
plt.scatter(X, Y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter plot of X vs Y (Quadratic Data)')
plt.grid()
plt.show() 
```

```python
Y = (Y - np.min(Y)) / (np.max(Y) - np.min(Y))
plt.figure(figsize=(8, 6))
plt.scatter(X, Y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter plot of X vs Y (Quadratic Data)')
plt.grid()
plt.show() 
```

```python

model = Sequential([
    Dense(30, input_shape=(1, ), activation='tanh'),
    Dense(30, activation='tanh'),
    Dense(1, activation=None)   # Output layer for regression
])
model.compile(optimizer='sgd', loss='mean_squared_error')
model.fit(X, Y, epochs=100, verbose=0)

loss_values = model.history.history['loss']
plt.figure(figsize=(8, 6))
plt.plot(loss_values)
plt.xlabel('Epoch') 
plt.ylabel('Loss')
plt.title('Training Loss Over Epochs for Deep NN')
plt.grid()
plt.show()
```

```python
Y_pred = model.predict(X)
plt.figure(figsize=(8, 6))
plt.scatter(X, Y, label='Data', color='blue')
plt.scatter(X, Y_pred, label='Model Prediction', color='red')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter plot of X vs Y with Deep NN Model Prediction')
plt.legend()
plt.grid()
plt.show()
```

```python
# Đưa Y_pred từ [0, 1] về lại scale gốc của Y
Y = 2.0 * X ** 2 - 3.0 * X + 5.0 + np.random.normal(0, 20, 500)
Y_pred_scale_upd = Y_pred * (np.max(Y) - np.min(Y)) + np.min(Y)
plt.figure(figsize=(8, 6))


plt.scatter(X, Y, label='Data', color='blue')
plt.scatter(X, Y_pred_scale_upd, label='Model Prediction', color='red')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter plot of X vs Y with Deep NN Model Prediction (Scaled Back)')
plt.legend()
plt.grid()
plt.show()
```
