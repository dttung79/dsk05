```python
import pandas as pd
```

```python
movies_df = pd.read_csv('movies.csv')
# check NA
print(movies_df.isna().sum())
```

```python
ratings_df = pd.read_csv('ratings.csv')
# check NA
print(ratings_df.isna().sum())
```

```python
print(movies_df.shape)
print(ratings_df.shape)
```

```python
movies_df['year'] = movies_df['title'].str.extract(r'\((\d{4})\)')
```

```python
movies_df.head()
```

```python
movies_df['title'] = movies_df['title'].str.replace(r'\(\d{4}\)', '', regex=True).str.strip()
movies_df.head()
```

```python
genres = 'Adventure|Animation|Children|Comedy|Fantasy'
genres = genres.split('|')
print(genres)
```

```python
movies_df['genres'] = movies_df['genres'].str.split('|')
movies_df.head()
```

```python
genres_cols = []
new_movies_df = movies_df.copy(deep=True)
for index, row in movies_df.iterrows():
    genres_cols.append(index)
    for g in row['genres']:
        new_movies_df.at[index, g] = 1
    
new_movies_df.head()
```

```python
new_movies_df.fillna(0, inplace=True)
new_movies_df.head()
```

```python
ratings_df.drop('timestamp', axis=1, inplace=True)
ratings_df.head()
```

```python
tungdt_ratings = [
    {'title': 'Mission: Impossible', 'rating': 4.5},
    {'title': 'Final Destination', 'rating': 4.0},
    {'title': 'Superman', 'rating': 4.2},
    {'title': 'GoldenEye', 'rating': 4.3},
    {'title': 'Mortal Kombat', 'rating': 3.5},
    {'title': 'Death Race', 'rating': 4.0}
]

tungdt_ratings = pd.DataFrame(tungdt_ratings)
print(tungdt_ratings)
```

```python
tungdt_ratings_id = movies_df[movies_df['title'].isin(tungdt_ratings['title'])]
tungdt_ratings = pd.merge(tungdt_ratings_id, tungdt_ratings, on='title')
print(tungdt_ratings)
```

```python
print(new_movies_df.head())
```

```python
print(tungdt_ratings.head())
```

```python
tungdt_ratings_genres = new_movies_df[new_movies_df.movieId.isin(tungdt_ratings.movieId)]
print(tungdt_ratings_genres)
```

```python
tungdt_ratings.drop(columns=['genres', 'year'], inplace=True)
```

```python
print(tungdt_ratings)
```

```python
tungdt_ratings_genres.drop(columns=['movieId','title', 'genres', 'year'], inplace=True)
tungdt_ratings_genres.reset_index(drop=True, inplace=True)
print(tungdt_ratings_genres)
```

```python
tungdt_ratings_genres.reset_index(drop=True, inplace=True)
print(tungdt_ratings_genres)
```

```python
print(tungdt_ratings.shape)
print(tungdt_ratings_genres.shape)
```

```python
tungdt_profile = tungdt_ratings_genres.T.dot(tungdt_ratings['rating'])
print(tungdt_profile)
```

```python
new_movies_df = new_movies_df.set_index('movieId')
print(new_movies_df.head())
```

```python
new_movies_df = new_movies_df.drop(columns=['title', 'genres', 'year'])
print(new_movies_df.head())
```

```python
recommend_tungdt = (new_movies_df.dot(tungdt_profile)) / tungdt_profile.sum()
print(recommend_tungdt)
```

```python
recommend_tungdt.sort_values(ascending=False, inplace=True)
print(recommend_tungdt.head(10))
```

```python
print(movies_df.head())
movies_df_reset = movies_df.set_index('movieId', drop=True)
print(movies_df_reset.head())
```

```python
top_recommend_index = recommend_tungdt.index[:10]
top_recommend_movies = movies_df_reset.loc[top_recommend_index.tolist(), :]
print(top_recommend_movies[['title', 'genres']])
```
